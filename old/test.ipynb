```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import numpy as np
import datasets
```

```python
base_model_name ="xlm-roberta-base"
model_name = "/scratch/project_2009199/pytorch-registerlabeling/models/xlm-roberta-large/labels_upper/en-fi-fr-sv-tr_en-fi-fr-sv-tr/seed_42/fold_1"
```

```python
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(base_model_name)
```

```python
text = "Moday 20th October a bad car crash on highway 7. More at..."
```

```python
out = model(**tokenizer(text, return_tensors='pt', truncation='only_first'), output_hidden_states=True)
```

```python
print(out)
```

```python
num_layers = len(out["hidden_states"])
indices = np.array([0, num_layers//2, -1], dtype=int)
print(indices)
print(out["hidden_states"][-1].size())
```

```python
embed = [torch.mean(out["hidden_states"][i],axis=1).cpu() for i in indices]
print(embed)
print(embed[0].size())
```
